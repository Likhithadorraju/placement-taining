```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Load the datasets
train_data = pd.read_csv('/content/data/train Data.csv')  # Replace with correct path
test_data = pd.read_csv('/content/data/test Data.csv')    # Replace with correct path

# Step 1: Select numeric columns for training and test data
train_data_numeric = train_data.select_dtypes(include=['float64', 'int64'])
test_data_numeric = test_data.select_dtypes(include=['float64', 'int64'])

# Step 2: Align columns between training and test data
train_numeric_cols = train_data_numeric.columns
test_numeric_cols = test_data_numeric.columns

# If there are missing columns in test data, add them with NaN values
missing_cols = set(train_numeric_cols) - set(test_numeric_cols)
for col in missing_cols:
    test_data_numeric[col] = float('nan')  # Add missing columns in test data

# If there are extra columns in the test data, drop them
extra_cols = set(test_numeric_cols) - set(train_numeric_cols)
test_data_numeric = test_data_numeric.drop(columns=extra_cols, errors='ignore')  # Ignore errors if columns are already dropped

# Ensure that both train and test data have the same columns in the same order
test_data_numeric = test_data_numeric[train_numeric_cols]

# Step 3: Imputation - Handle missing values in both training and test data
imputer = SimpleImputer(strategy='mean')  # You can change strategy to 'median' or 'most_frequent'

# Fit the imputer on the training data
imputer.fit(train_data_numeric)

# Get columns after imputer transform to avoid shape mismatch:
train_data_numeric_imputed_cols = train_data_numeric.columns[~train_data_numeric.isnull().all()]

# Step 4: Impute missing values in both training and test data
train_data_numeric_imputed = pd.DataFrame(imputer.transform(train_data_numeric), columns=train_data_numeric_imputed_cols) # Use the determined columns
test_data_numeric_imputed = pd.DataFrame(imputer.transform(test_data_numeric), columns=train_data_numeric_imputed_cols) # Use the determined columns

# Step 5: Scaling - Standardize the numeric columns
scaler = StandardScaler()

# Fit the scaler on the training data
train_data_scaled = scaler.fit_transform(train_data_numeric_imputed)

# Transform the test data using the same scaler
test_data_scaled = scaler.transform(test_data_numeric_imputed)

# Step 6: Optional - Check the first few rows of the scaled data
print("Scaled Training Data (First 5 rows):")
print(train_data_scaled[:5])

print("\nScaled Test Data (First 5 rows):")
print(test_data_scaled[:5])
```

```python
from sklearn.mixture import GaussianMixture

# Fit Gaussian Mixture Model
gmm_model = GaussianMixture(n_components=3)  # Adjust the number of components (clusters)
train_data['gmm_cluster'] = gmm_model.fit_predict(train_data_scaled)

# Visualize the clusters
sns.scatterplot(x=train_data['feature1'], y=train_data['feature2'], hue=train_data['gmm_cluster'], palette="Set2")
plt.title("Gaussian Mixture Model Clustering")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Apply Agglomerative Clustering
hierarchical_clustering = AgglomerativeClustering(n_clusters=3)  # Choose the number of clusters
train_data['hierarchical_cluster'] = hierarchical_clustering.fit_predict(train_data_scaled)

# Visualize or use these cluster labels for further analysis
print(train_data[['id', 'hierarchical_cluster']].head())
```

```python
from sklearn.mixture import GaussianMixture

# Apply GMM with 3 clusters (adjust as needed)
gmm = GaussianMixture(n_components=3)
gmm_clusters = gmm.fit_predict(train_data_scaled)

# Add the GMM cluster labels to the training data
train_data['gmm_cluster'] = gmm_clusters

# Visualize or use the GMM clusters for further analysis
print(train_data[['id', 'gmm_cluster']].head())
```
